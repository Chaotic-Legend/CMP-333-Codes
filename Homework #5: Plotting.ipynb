1. Load the CSV file as a pandas data frame. Display its first 5 rows.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cereal.csv")
display(df.head())
```

2. Create a scatter plot to show the relationship between "fiber" and "rating". Make sure the figure has a title, and both axes are labeled.

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'figure.autolayout': True})
plt.figure(figsize=(8,6))
plt.scatter(df["fiber"], df["rating"], color="#1f77b4", edgecolors="black", alpha=0.8)
plt.title("Relationship Between Fiber and Rating", fontsize=12, weight='bold')
plt.xlabel("Fiber")
plt.ylabel("Rating")
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

3. Count the number of products made by each manufacturer (use the 'mfr' column). Create a bar plot to show these numbers.

```python
manufacturer_counts = df["mfr"].value_counts()
plt.figure(figsize=(8,6))
bars = plt.bar(manufacturer_counts.index, manufacturer_counts.values, color='orange', edgecolor='black')
plt.title("Number of Cereals by Manufacturer", fontsize=12, weight='bold')
plt.xlabel("Manufacturer")
plt.ylabel("Number of Products")
plt.xticks(rotation=0)
plt.grid(True, linestyle=':', alpha=0.7)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
             str(bar.get_height()), ha='center', va='bottom', fontsize=10, weight='bold')
plt.show()
```

4. Create a histogram of the ratings. Use 8 bins instead of the default 10 bins.

```python
plt.figure(figsize=(8,6))
counts, bins, patches = plt.hist(df["rating"], bins=8, color='lightgreen', edgecolor='black', alpha=0.9)
for count, patch in zip(counts, patches):
    plt.text(patch.get_x() + patch.get_width()/2, count + 0.1,
             int(count), ha='center', va='bottom', fontsize=10, weight='bold')
plt.title("Distribution of Cereal Ratings", fontsize=12, weight='bold')
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

5. Create a pie chart to show the number of cold cereals and hot cereals (use the 'type' column).

```python
type_counts = df["type"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    type_counts,
    labels=type_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=["#66b3ff", "#ff6666"],
    wedgeprops={'edgecolor': 'black'})
plt.title("Proportion of Cold vs. Hot Cereals", fontsize=12, weight='bold')
plt.show()
```

6. Create a boxplot for all the numerical features.

```python
plt.figure(figsize=(10,6))
df.select_dtypes(include='number').boxplot(patch_artist=True)
plt.title("Boxplot of Numerical Features", fontsize=12, weight='bold')
plt.xticks(rotation=45)
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
